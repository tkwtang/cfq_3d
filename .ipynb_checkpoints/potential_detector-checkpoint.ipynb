```python
def get_XYU(domain = [-4, 4], resolution = 50, detail_info = False):
    """
    To get the 2D data of the potential for contour plots
    """
    x_vec = np.linspace(domain[0], domain[1], resolution)
    x, y = np.meshgrid(x_vec, x_vec)
    U_12 = coupled_fq_pot.potential(X, Y, 0, 0, 0, 0, params = initial_parameter_dict.values())
    U_23 = coupled_fq_pot.potential(0, X, Y, 0, 0, 0, params = initial_parameter_dict.values())
    U_13 = coupled_fq_pot.potential(X, 0, Y, 0, 0, 0, params = initial_parameter_dict.values())

    
    return X, Y, U_12, U_23, U_13
```
